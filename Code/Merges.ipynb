```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tech = pd.read_csv('/Users/annamezhlauk/Desktop/ECO225Project/Data/Tech_salaries.csv')
objects = pd.read_csv('/Users/annamezhlauk/Desktop/ECO225Project/Data/Startups/objects.csv', low_memory = False)
print(tech)
```

```python
tech = tech.drop([44, 53, 267, 281, 522, 586, 646, 789, 1303, 1422, 1423, 1424, 1555, 1595, 1640, 1642], axis=0)
# dropping observations with "spam" position names



tech = tech.drop([1448, 1132, 860, 766, 689, 677, 387, 21], axis=0)
# dropping observations where compensation isn't in USD

tech = tech.reset_index(drop=True)

print(tech)
```

```python
tech.head()

tech = tech[(tech['annual_base_pay'] >= 2000) & (tech['annual_base_pay'] <= 5000000)]
tech = tech.reset_index(drop=True)

print(tech)
```

```python
objects.rename(columns = {"normalized_name": "employer_name"}, inplace = True)
print(objects.dtypes)
objects.head(25)
```

```python
filtered_for_funding = objects[objects['funding_total_usd'] > 5000].dropna(subset=['funding_total_usd'])

plt.xscale('log')
sns.histplot(filtered_for_funding['funding_total_usd'], kde=False)
plt.title('Startup Total Funding Distribution')
plt.xlabel('Total Funding in USD')
plt.ylabel('Frequency')
plt.show()


filtered_for_funding.describe()
```

```python
intersection = combo = tech.merge(objects, how="inner", on = "employer_name")
intersection.describe()
```

```python
filtered_intersection = intersection[
    (intersection['funding_total_usd'] > 1000) & 
    (intersection['job_title'].str.lower() == 'software engineer')
    ]

plt.figure(figsize=(10, 6))
sns.scatterplot(x=filtered_intersection['funding_total_usd'], y=filtered_intersection['annual_base_pay'], alpha=0.5)
plt.xscale("log")
plt.xlabel("Total Funding for Startups Funded by More Than $1,000")
plt.ylabel("Annual Base Pay (USD)")
plt.title("Total Startup Funding vs. Salary for Software Engineers")
plt.show()
```

```python
plt.yscale("log")

sns.boxplot(x=intersection['entity_type'], y=intersection['annual_base_pay'])
plt.xlabel("Startup Entity Type")
plt.ylabel("Annual Base Pay (USD)")
plt.title("Salary Distribution by Startup Entity Type")
plt.xticks(rotation=45)
plt.show()
```

```python
print(intersection['status'].unique())
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=intersection['status'], 
            y=intersection['annual_base_pay'], 
            estimator=lambda x: sum(x) / len(x))
plt.xlabel("")
plt.ylabel("Annual Base Pay (USD)")
plt.title("Average Employee Salary by Startup Status")
plt.xticks(rotation=45) 
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=intersection['funding_rounds'], y=intersection['annual_base_pay'])
plt.xlabel("Number of Funding Rounds")
plt.ylabel("Annual Base Pay (USD)")
plt.title("Salary Trends Across Funding Rounds")
plt.show()
```
