```python
!pip install -q pandas matplotlib seaborn numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
tech_salaries = pd.read_csv('/Users/annamezhlauk/Desktop/ECO225Project/Data/Tech_salaries.csv')
startups_main = pd.read_csv('/Users/annamezhlauk/Desktop/ECO225Project/Data/Startups/objects.csv', low_memory = False)
```

```python
# dropping observations with "spam" position names
tech_salaries = tech_salaries.drop([44, 53, 267, 281, 522, 586, 646, 789, 1303, 1422, 1423, 1424, 1555, 1595, 1640, 1642], axis=0)


# dropping observations where compensation isn't in USD
tech_salaries = tech_salaries.drop([1448, 1132, 860, 766, 689, 677, 387, 21], axis=0)

# dropping improbable salaries that could be data entry mistakes
tech_salaries = tech_salaries[(tech_salaries['annual_base_pay'] >= 5000) & (tech_salaries['annual_base_pay'] <= 3000000)].reset_index(drop=True)
```

```python
startups_main.rename(columns = {"normalized_name": "employer_name"}, inplace = True)
startups_main.rename(columns = {"entity_id": "startup_id"}, inplace = True)
```

```python
startups_main['startup_id'] = startups_main['startup_id'].astype('Int64')
startups_and_salaries = tech_salaries.merge(startups_main, how="left", on="employer_name")
```

```python
# making a column for the number of employees in the data for each startup
startups_and_salaries['num_employees'] = startups_and_salaries.groupby('startup_id')['salary_id'].transform('count')
```

```python

def haversine(lat1, lon1, lat2, lon2):
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    radius_earth_miles = 3960  # Radius of the Earth in miles
    return radius_earth_miles * c


def count_competitor_employees_v3(df, radius):
    # Initialize competitors column
    df['competitors'] = np.nan  # Default to NaN

    # Iterate over unique job categories
    for category in df['job_title_category'].dropna().unique():
        if category == "Other":
            continue  # Skip processing for "Other"

        category_df = df[df['job_title_category'] == category]

        # Iterate over each startup
        for i, row in category_df.iterrows():
            if pd.isna(row['location_latitude']) or pd.isna(row['location_longitude']):
                continue  # Skip missing location data
            
            lat1, lon1 = row['location_latitude'], row['location_longitude']
            competitors = 0

            # Compare against other startups in the same category
            for j, comp_row in category_df.iterrows():
                if pd.notna(row['startup_id']) and pd.notna(comp_row['startup_id']) and row['startup_id'] != comp_row['startup_id']:
                    if pd.notna(comp_row['location_latitude']) and pd.notna(comp_row['location_longitude']):
                        lat2, lon2 = comp_row['location_latitude'], comp_row['location_longitude']
                        if haversine(lat1, lon1, lat2, lon2) <= radius:
                            competitors += comp_row['num_employees']

            df.at[row.name, 'competitors'] = competitors

    return df
```

```python
test = count_competitor_employees_v3(startups_and_salaries, 300)
pd.set_option('display.max_columns', None)
test.head()
```

```python
# Set style
#sns.set_style("whitegrid")

# Filter out the categories you don't want
filtered_data = test[test['status'].isin(['operating', 'ipo', 'acquired'])]

# Compute the average annual base pay for each startup
startup_avg_pay = filtered_data.groupby(['startup_id', 'status'])['annual_base_pay'].mean().reset_index()

# Count the number of startups in each status category
status_counts = startup_avg_pay['status'].value_counts()

# Create the sample size table
sample_size_table = pd.DataFrame(status_counts).reset_index()
sample_size_table.columns = ['Status', 'Sample Size']

# Create the boxplot
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='status', y='annual_base_pay', data=startup_avg_pay, order=status_counts.index)
plt.title("Average Annual Base Pay Across Startup Status Categories")
plt.ylabel("Average Annual Base Pay (USD)")
plt.xlabel("Startup Status")
plt.yscale("log")  # Log scale for salary distribution


# Add sample size annotations (number of startups per category)
for i, status in enumerate(status_counts.index):
    count = status_counts[status]
    plt.text(i, ax.get_ylim()[0], f'n={count}', ha='center', va='bottom', fontsize=12, color='black')

# Add a note at the bottom of the plot (below the labels)
plt.figtext(0.5, -0.05, 
            "Note: The 'Live', 'Private', 'Closed', and 'Beta' categories were omitted because their sample sizes were too small to analyze. "
            "Sample sizes: Live=5, Private=3, Closed=2, Beta=1.", 
            ha='center', va='center', fontsize=10, color='black')

plt.tight_layout()  # Ensure everything fits within the plot area
plt.show()

# 3. Histogram of number of employees at a startup
plt.figure(figsize=(8, 6))
sns.histplot(test['num_employees'].dropna(), bins=25)
plt.title("Distribution of Number of Employees at Startups")
plt.xlabel("Number of Employees")
plt.ylabel("Count")
plt.show()


# 5. Bar graph of number of competitors for all numbers of funding rounds
plt.figure(figsize=(10, 6))
sns.barplot(x='funding_rounds', y='competitors', data=test, errorbar=None)
plt.title("Number of Competitors vs. Funding Rounds")
plt.xlabel("Number of Funding Rounds")
plt.ylabel("Number of Competitors")
plt.show()

# 6. Bar graph of annual base pay for all numbers of funding rounds
plt.figure(figsize=(10, 6))
sns.barplot(x='funding_rounds', y='annual_base_pay', data=test, errorbar=None)
plt.title("Annual Base Pay vs. Funding Rounds")
plt.xlabel("Number of Funding Rounds")
plt.ylabel("Annual Base Pay (USD)")
plt.yscale("log")  # Log scale if needed
plt.show()

# 7. Scatter plot of funding level vs. wage premium

# Group by startup and calculate average experience for employees
startup_experience = test.groupby('startup_id').agg(
    avg_experience=('total_experience_years', 'mean'),
    total_funding=('funding_total_usd', 'first')  # Assuming funding is constant per startup
).reset_index()

# Scatterplot of average experience vs. total funding
plt.figure(figsize=(8, 6))
sns.scatterplot(x='avg_experience', y='total_funding', data=startup_experience, alpha=0.5)
plt.title("Average Employee Experience vs. Startup Funding")
plt.xlabel("Average Employee Experience (Years)")
plt.ylabel("Total Funding (USD)")
plt.yscale("log")  # Log scale for funding amounts
plt.show()

# Create a histogram of the number of competitors
plt.figure(figsize=(8, 6))
sns.histplot(test['competitors'], bins=30, kde=False, color='blue')

# Customize the plot
plt.title("Distribution of Number of Competitors")
plt.xlabel("Number of Competitors")
plt.ylabel("Frequency")
plt.show()
```

```python
# Define IPO vs Operating based on the 'status' column
startups_and_salaries['ipo_status'] = startups_and_salaries['status'].apply(
    lambda x: 'IPO' if x == 'ipo' else 'Operating' if x == 'operating' else 'Other'
)

# Filter out the 'Other' status for the plot
filtered_data = startups_and_salaries[startups_and_salaries['ipo_status'] != 'Other']

# Group by 'ipo_status' and calculate average salary for each group at different levels of experience
grouped_data = filtered_data.groupby(['total_experience_years', 'ipo_status']).agg({'annual_base_pay': 'mean'}).reset_index()

# Plotting the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='total_experience_years', y='annual_base_pay', hue='ipo_status', data=grouped_data, palette='viridis', linewidth=2, legend=False)

# Add labels for each line
for line, ipo_status in zip(plt.gca().lines, ['IPO', 'Operating']):
    y_position = line.get_ydata()[-1]  # Get the last y value for annotation
    plt.text(line.get_xdata()[-1], y_position, ipo_status, color=line.get_color(), 
             ha='left', va='center', fontsize=12)

plt.title('Salary vs. Experience by IPO vs Operating')
plt.xlabel('Years of Experience')
plt.ylabel('Average Annual Base Pay (USD)')
plt.tight_layout()
plt.show()

```
