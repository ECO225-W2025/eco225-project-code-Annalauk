```python
tech_salaries = pd.read_csv('/Users/annamezhlauk/Desktop/ECO225Project/Data/Tech_salaries.csv')
startups_main = pd.read_csv('/Users/annamezhlauk/Desktop/ECO225Project/Data/Startups/objects.csv', low_memory = False)
```

```python
# dropping observations with "spam" position names
tech_salaries = tech_salaries.drop([44, 53, 267, 281, 522, 586, 646, 789, 1303, 1422, 1423, 1424, 1555, 1595, 1640, 1642], axis=0)


# dropping observations where compensation isn't in USD
tech_salaries = tech_salaries.drop([1448, 1132, 860, 766, 689, 677, 387, 21], axis=0)

# dropping improbable salaries that could be data entry mistakes
tech_salaries = tech_salaries[(tech_salaries['annual_base_pay'] >= 5000) & (tech_salaries['annual_base_pay'] <= 4000000)].reset_index(drop=True)
```

```python
startups_main.rename(columns = {"normalized_name": "employer_name"}, inplace = True)
startups_main.rename(columns = {"entity_id": "startup_id"}, inplace = True)
```

```python
startups_main['startup_id'] = startups_main['startup_id'].astype('Int64')
```

```python
startups_and_salaries = tech_salaries.merge(startups_main, how="left", on="employer_name")
#startups_and_salaries_clean = startups_and_salaries.dropna(subset=['startup_id', 'job_title_category'])
```

```python
def count_competitor_employees_v2(df):
    # Group by both job_title_category, startup_id, and location_country to count employees in each job title category
    job_counts = df.groupby(['job_title_category', 'startup_id', 'location_country']).size().reset_index(name='employee_count')

    # Calculate total employees in each job title category for the same location_country
    total_job_counts = df.groupby(['job_title_category', 'location_country']).size().reset_index(name='total_in_category')

    # Merge the total count with the job_counts on job_title_category and location_country
    job_counts = job_counts.merge(total_job_counts, on=['job_title_category', 'location_country'], how='left')

    # Calculate competitors (total employees in the category in the same location - current startup's employees)
    job_counts['competitors'] = job_counts['total_in_category'] - job_counts['employee_count']

    # Create a new DataFrame indexed by startup_id with the competitors count
    competitor_df = job_counts[['startup_id', 'competitors']].drop_duplicates(subset='startup_id')

    # Set the index to 'startup_id'
    competitor_df.set_index('startup_id', inplace=True)

    # Merge this back to the original DataFrame
    df = df.merge(competitor_df, on='startup_id', how='left')

    return df
```

```python
test = count_competitor_employees_v2(startups_and_salaries)
test.head()
```

```python
test.describe()
```

```python
# Check if the 'number_competitor_employees' column has the same value for each 'startup_id'
check_values = test.groupby('startup_id')['competitors'].nunique()

# Display startup_ids that have more than one unique value
duplicates = check_values[check_values > 1]

# Print duplicates if any
print(duplicates)
```

```python
# Count the number of employees per startup_id
test['num_employees'] = test.groupby('startup_id')['salary_id'].transform('count')
# Compute the average salary per startup_id
test['avg_salary'] = test.groupby('startup_id')['annual_base_pay'].transform('mean')
```

```python

# Create a scatter plot to visualize the relationship between avg_salary and competitors
plt.figure(figsize=(10, 6))
sns.scatterplot(data=test, x='competitors', y='avg_salary', alpha=0.7)

# Add titles and labels
plt.title('Relationship between Average Salary and Competitors', fontsize=16)
plt.xlabel('Number of Competitors', fontsize=12)
plt.ylabel('Average Salary (USD)', fontsize=12)

# Display the plot
plt.show()
```
