## INTRODUCTION

The study of startups is a compelling research area for those concerned with innovation and markets. Startups have great potential to introduce novel technologies, but proper strategy is critical to achieving the brand recognition and reputation necessary for these impacts (Hashai & Markovich, 2017). Moreover, the positive effects aren’t limited to technology development. Startups can also be a source of attractive jobs by offering interesting positions or a close-knit work environment. The data surrounding startup pay suggests lower wages in general, but without a clear consensus. First, based on Danish registry data, Sorenson et al. concludes that startup employees earn about 17% less than employees at larger comparable firms over 10 years (Sorenson et al., 2021). Estimates from Germany also support lower salaries when everything else is held constant (Brixy et al., 2006). However, J. Daniel Kim’s work finds a wage premium for venture capital backed firms due to their selection of especially skillful employees (Kim, 2018). 

Interestingly, a study on MIT graduates finds no wage differences between startups and estbalished companies when holding worker ability and skills constant (Kim, 2016). Although this analysis may not be externally valid for workers of lower ability, it emphasizes the importance of measures of employee experience and skill in any investigation of wage determination. 

Now, in regards to wage determination within startups, Li Yang’s research on market competition and employee wage for Chinese companies found a statistically significant positive effect of elevated competition (Yang, 2016). This work is relatively unique in studying wage determination for all employees rather than founders and it isolates the impact of competition itself on wages. Consequently, it directly motivates my own analysis. In this paper, I examine the relationship between an estimator for startup hiring-related competition and wages by merging startup data from 2013 with technology sector salary data from 2016. Although I detect a very small effect of competition by itself, I explore differences in incentives for high pay at acquired startups (in particular realted to hiring strategy and employee experience level), analyze regional wage and compettition differences in the US, and find a significant funding level effect on wage after controlling for startup status.

## DATA EXPLANATION

I use individual level salary survey responses from Hacker News, combined with entity level startup investment data from Crunchbase. Observations are merged on company name, so startups are grouped with corresponding 2016 salaries of their employees.

The main outcome variable is annual base pay in USD from the salary survey responses.

### Explanatory variables:

*Total startup funding* (USD) is examined since it can act as a potential measure of the ability of a startup to pay high salaries. Comparing the impact of the number of competitor employees only for startups with similar funding can better isolate the effect.

Similarly, the *number of funding rounds* indicates the level of funding available, but also reveals how established the startup is. A later stage startup is likely to have received more funding rounds since it has proven some level of reputation and success to investors. As a result, it is another critical control variable.

A more explicit measure of how established a startup is its official status. *Startup status* is a a categorical variable with 7 unique values: acquired, beta, closed, IPO, live, operating, and private. Startups in different stages are likely going to have different sensitivities to competitors. For example, a closed or acquired startup may not be changing wages at all in response to competition since it has potentially recently undergone major structural changes that diverted its attention.

An employee’s *level of experience* is known to be correlated with wage. Including this variable clarifies whether the potential difference in wage for startups with a greater prevalence of competitor employees is due to competition pressure directly or different employee selection tied to experience.

The *number of employees* in the data for each startup provides another potential dimension that can be impacted by the level of competitor employees. Instead of later hiring employees at a higher wage, perhaps startups anticipating facing competition strategically hire more workers ahead of time.

The *number of competitors* is a metric which estimates hiring competition by counting the number of competitor employees (meaning other workers in the same industry but not the same startup) within a 300 mile radius of each employee. This is a useful measure of hiring competition because firms that hire workers in the same industry and the same location are competitors for labor. As a result, increased competition should theoretically be associated with higher wages.

### Helper functions for variable creation:

```python
def distance_formula(lat1, lon1, lat2, lon2):
    """Return the distance between the points at lat1 lon1 and lat2 lon2 in miles.
    """
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # this is the haversine formula (from community.esri.com)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    
    radius_earth_miles = 3960
    return radius_earth_miles * c


def count_competitor_employees(df, radius):

    df['competitors'] = np.nan

    # this is to make sure employees within the 'other' category aren't counted as competitor employees
    for category in df['job_title_category'].dropna().unique():
        if category == "Other":
            continue

        category_df = df[df['job_title_category'] == category]

        # iterates over each startup & exits the loop early if location data is missing
        for i, row in category_df.iterrows():
            if pd.isna(row['location_latitude']) or pd.isna(row['location_longitude']):
                continue
            
            lat1, lon1 = row['location_latitude'], row['location_longitude']
            competitors = 0

            
            # compares with startups in the same category
            for j, comp_row in category_df.iterrows():
                
                # checking for missing values
                if pd.notna(row['startup_id']) and pd.notna(comp_row['startup_id']) and row['startup_id'] != comp_row['startup_id']:
                    if pd.notna(comp_row['location_latitude']) and pd.notna(comp_row['location_longitude']):
                        lat2, lon2 = comp_row['location_latitude'], comp_row['location_longitude']

                        # adds competitors based on distance
                        if distance_formula(lat1, lon1, lat2, lon2) <= radius:
                            competitors += comp_row['num_employees']

            df.at[row.name, 'competitors'] = competitors

    return df
```

## REGRESSIONS

```python
y = startups_and_salaries_clean['annual_base_pay']

# data cleaning function for regression
def clean_data(X):
    # converting to numeric & coercing errors into NaN
    X = X.apply(pd.to_numeric, errors='coerce')
    
    X = X.dropna()
    return X

# specification 1
X1 = startups_and_salaries_clean[['competitors']]
X1 = sm.add_constant(X1)
X1 = clean_data(X1) 


# specification 2
X2 = startups_and_salaries_clean[['competitors', 'funding_rounds']]
X2 = sm.add_constant(X2)
X2 = clean_data(X2) 

# specification 3
startups_and_salaries_clean.loc[:, 'any_funding'] = (startups_and_salaries_clean['funding_rounds'] > 0).astype(int)
X3 = startups_and_salaries_clean[['competitors', 'any_funding']] 
X3 = sm.add_constant(X3) 
X3 = clean_data(X3)  
X3 = X3.astype(float)  
X3 = X3.apply(pd.to_numeric, errors="coerce")  


# specification 4
status_dummies = pd.get_dummies(startups_and_salaries_clean['status'])
status_dummies = status_dummies.drop(columns=['beta'])
X4 = startups_and_salaries_clean[['competitors']].join(status_dummies)
X4 = sm.add_constant(X4) 
X4 = clean_data(X4)
X4 = X4.astype(float)
X4 = X4.apply(pd.to_numeric, errors="coerce")

# specification 5
status_dummies = pd.get_dummies(startups_and_salaries_clean['status'])
status_dummies = status_dummies.drop(columns=['beta'])
X5 = startups_and_salaries_clean[['competitors', 'funding_rounds']].join(status_dummies)
X5 = sm.add_constant(X5) 
X5 = clean_data(X5)
X5 = X5.astype(float)
X5 = X5.apply(pd.to_numeric, errors="coerce")


# specification 6
X6 = startups_and_salaries_clean[['competitors', 'total_experience_years']]
X6 = sm.add_constant(X6)
X6 = clean_data(X6) 

model1 = sm.OLS(y.loc[X1.index], X1).fit()
model2 = sm.OLS(y.loc[X2.index], X2).fit()
model3 = sm.OLS(y.loc[X3.index], X3).fit()
model4 = sm.OLS(y.loc[X4.index], X4).fit()
model5 = sm.OLS(y.loc[X5.index], X5).fit()
model6 = sm.OLS(y.loc[X6.index], X6).fit()
model7 = sm.OLS(y.loc[X7.index], X7).fit()

# specification 7
X7 = startups_and_salaries_clean[['competitors', 'num_employees']]
X7 = sm.add_constant(X7)
X7 = clean_data(X7) 

custom_order = ["competitors", "funding_rounds", "any_funding", "live", "operating", "ipo", "acquired", "closed",
                "total_experience_years", "num_employees", "const"]
stargazer = Stargazer([model1, model2, model3, model4, model5, model6, model7])
stargazer.covariate_order(custom_order)
stargazer.title(r"""
<h3>Preferred Specification:</h3><br> 
\[
Y = \beta_0 + \beta_1 \text{competitors} + \beta_2 \text{funding\_rounds} + \beta_3 \text{live} 
+ \beta_4 \text{operating} + \beta_5 \text{ipo} + \beta_6 \text{acquired} + \beta_6 \text{closed} + \epsilon
\]
""")
display(HTML(stargazer.render_html()))
```

From specification 1, we see that a greater number of competitor employees is associated with only slightly higher salaries. The effect is statistically significant at a 1% level (highly statistically significant) and it small but notable. For a startup employee, having 100 more competitor employees within a 300 mile radius is associated with a salary roughly $1860 higher on average, a precise null result. 

After adding funding rounds to the regression, we notice that the estimate of this effect falls, indicating that startups with more competitors also tended to have more funding rounds. Funding rounds is particularly interesting as a variable since it signals both how developed a startup is and how much funding may be available to spend on employee wages. 

In specification 1, too much of the variation in salary was attributed to varying numbers of competitors when funding rounds was actually explaining some of it. For each additional funding round, employee salaries were about $2400 higher on average when controlling for number of competitor employees.

The funding indicator variable in specification 3 did not contribute to the regression, but this is likely due to many firms having zero funding, limiting statistical precision since the sample size for funded startups is small in comparison.

Now, incorperating startup status indicators in specification 4, we notice that only the "acquired" indicator had an effect. When holding competitor employees constant, acquired startups paid lower salaries than beta startups by roughly $52,000. This difference is statistically significant at the 5% level. While beta startups may need to pay relatively attractive wages to attract talent, acquired startups have undergone administrative changes that likely have an impact on compensation. For example, layoffs after an acquisition can result in lower salaries on average since any new employees hired can be paid less than the previous workers.

Moving on to the preferred specification (specification 5), we notice that controlling for funding rounds increases this effect. This implies that acquired startups had a higher number of funding rounds than beta startups, muting the effect of status on employee salary. Funding rounds had a greater and more statistically significant effect than in specification 2. When holding constant startup status and number of competitor employees, each additional funding round was associated with employee wages about $2600 higher on average. Additionally, the coefficient on the number of competitor employees decreases further, serving as more evidence of a marginal salary effect at most. This specification in particular contributes to the research question. We had theorized that IPO startups paid higher salaries using earlier summary statistics, when in reality, the only status-specific difference is between acquired and beta startups when comparing firms with the same number of funding rounds and competitor employees.

Specification 6 shows that an extra year of experience is associated with a salary roughly $2800 higher on average among firms with the same number of competitor employees. The coefficient on competitors also increases, indicating that firms who had higher levels of comeptitor employees also had less experienced employees on average. This result suggests that startups may prefer less experience when they are facing hiring competition. However, we did see evidence of several especially large firms having many competitors in project 1. It could be that those big firms are simply hiring more inexperienced workers than average, inflating the regression coefficient.

Lastly, the coefficient on competitors decreases the most when controlling for the number of employees. Hiring 10 more employees is associated with employee wages that are about $3260 higher among firms with the same number of competitors. This means that these additional employees would be outside of a 300 mile radius of other employees, indicating a potential wage premium for firms hiring many remote workers.

```python
y = startups_and_salaries_clean['total_experience_years']

# data cleaning function for regression
def clean_data(X):
    X = X.apply(pd.to_numeric, errors='coerce')
    X = X.dropna()
    return X

status_dummies = pd.get_dummies(startups_and_salaries_clean['status'])
status_dummies = status_dummies.drop(columns=['beta'])

# specification 1
X1 = sm.add_constant(status_dummies)

# specification 2
X2 = startups_and_salaries_clean[['any_funding']].join(status_dummies)
X2 = sm.add_constant(X2)

# specification 3
X3 = startups_and_salaries_clean[['funding_rounds']].join(status_dummies)
X3 = sm.add_constant(X3)

# specification 4
X4 = startups_and_salaries_clean[['num_employees']].join(status_dummies)
X4 = sm.add_constant(X4)


X1 = X1.apply(lambda col: col.astype(int) if col.dtype == 'bool' else col)
X2 = X2.apply(lambda col: col.astype(int) if col.dtype == 'bool' else col)
X3 = X3.apply(lambda col: col.astype(int) if col.dtype == 'bool' else col)
X4 = X4.apply(lambda col: col.astype(int) if col.dtype == 'bool' else col)


X1 = clean_data(X1)
X2 = clean_data(X2)
X3 = clean_data(X3)
X4 = clean_data(X4)


y_aligned = y.loc[X1.index]


model1 = sm.OLS(y_aligned, X1).fit()
model2 = sm.OLS(y_aligned, X2).fit()
model3 = sm.OLS(y_aligned, X3).fit()
model4 = sm.OLS(y_aligned, X4).fit()

custom_order = ["live", "operating", "ipo", "acquired", "closed", "any_funding", "funding_rounds", "num_employees", "const"]


stargazer = Stargazer([model1, model2, model3, model4])
stargazer.covariate_order(custom_order)
stargazer.title(r"""
<h3>Preferred Specification:</h3><br> 
\[
Y = \beta_0 + \beta_1 \text{live} + \beta_2 \text{operating} + \beta_3 \text{ipo} 
+ \beta_4 \text{acquired} + \beta_5 \text{closed} + \beta_6 \text{num\_employees} + \epsilon
\]
""")
display(HTML(stargazer.render_html()))
```

In this set of regressions, we analyze the years of experience variable further. 

Immediately, we see from specification 1 that acquired startups are seemingly hiring less experienced employees on average. This would partly explain the lower wage estimates from earlier regressions. Acquired startups hire employees with about 3.8 fewer years of experience on average compared to beta startups. It's not that acquired startups are unfairly low wages necessarily, but they are hiring inexperienced workers who are generally paid less. 

Incorperating a funding indicator variable or a number of funding rounds variable does not contribute to the regression further. Changes in funding level do not seem to come with differences in salary when comparing startups of the same status. 

However, firms hiring 100 more employees have an average employee experience level that is roughly 2.3 years lower when comparing startups of the same status (from the coefficient in specification 4). Firms that have fewer employees tend to have very experienced ones, while firms with more employees hire less experienced workers on average. Although this effect is not huge, this regression is the preferred specification because it provides intuition for different hiring patterns. The number of competitor employees is corelated with the number of employees based on how this variable was calculated. However, the firms with very high competitor employees may be utlizing a different hiring strategy. These startups hire more employees on average, but these employees also have less experience. This makes the true effect from competition harder to estimate. It is clearer now why the coefficient on competitors changed the most when controlling for number of employees in the previous regression table.

Using 2013 Startup data from Crunchbase and 2016 technology sector salary data from Hacker News, this paper has investigated the relationship between a hiring competition metric and startup employee wages. Although I obtained precise null effect for number of competitor employees on startup employee wage when controlling for startup status, funding, employee experience, and number of employees, I provided a rigorous analysis of wage determining factors nonetheless. I observed similar wage patterns in California to McNeill and Aran's articles, supporting the idea of California as a startup haven. At the same time, San Francisco is seen as an area with subtely declining wages (McNeill, 2016). Additionally, my regression estimates present evidence of different hiring strategies followed by acquired startups then beta startups. Acquired startups were shown to pay less, but they also hired employees with less experience on average. Also, the firms that hired the most employees in the data had the lowest average experience levels. This relationship made it harder to preceisely measure the direct effect of hiring competition. Lastly, funding rounds were initially not found to be correlated with wage. However, it was later uncovered that each additional funding round raised employee salary by about $2600 on average among startups of the same status when controlling for competitior employees, a moderate effect. While competitor employees may not explain a lot of salary variation, this variable has still provided insights into wage setting incentives for startups of different statuses and funding levels.
